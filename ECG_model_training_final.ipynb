```python
import tensorflow as tf
import numpy as np
import os
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.applications import VGG16, ResNet50
from tensorflow.keras.applications.vgg16 import preprocess_input as vgg_preprocess
from tensorflow.keras.applications.resnet import preprocess_input as resnet_preprocess
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Seed for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

# Parameters
IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 16
NUM_CLASSES = 6
EPOCHS = 20
AUTO = tf.data.AUTOTUNE

def focal_loss(gamma=2., alpha=.25):
  """
    Focal Loss function.

    This function defines a custom loss function called Focal Loss, which is
    designed to address class imbalance in classification tasks. It modifies
    the standard cross-entropy loss by down-weighting the contribution of
    easy-to-classify examples and focusing more on hard examples.

    Args:
        gamma (float, optional): Focusing parameter. A higher value gives more
            weight to hard examples. Defaults to 2.0.
        alpha (float, optional): Weighting factor for the positive class.
            Defaults to 0.25.

    Returns:
        function: The focal_loss_fixed function, which calculates the focal loss.
    """
    def focal_loss_fixed(y_true, y_pred):
       # Clip predicted probabilities to avoid log(0) errors
        y_pred = tf.clip_by_value(y_pred, tf.keras.backend.epsilon(), 1 - tf.keras.backend.epsilon())
        # Calculate the standard cross-entropy loss
        cross_entropy = -y_true * tf.math.log(y_pred)
        # Calculate the modulating factor, which down-weights easy examples
        modulating_factor = tf.pow(1. - y_pred, gamma)
        # Calculate the alpha factor, which balances class weights
        alpha_factor = y_true * alpha + (1 - y_true) * (1 - alpha)
        # Calculate the Focal Loss
        focal_loss = alpha_factor * modulating_factor * cross_entropy
        # Return the mean Focal Loss over all examples
        return tf.reduce_mean(focal_loss)
    return focal_loss_fixed  # Return the inner function
## COllect balanced data from the dataset
def balanced_data_collection(data_dir):
    image_files = []
    labels = []
    class_names = []

    try:
      # Iterate over class folders
        class_folders = sorted([f for f in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, f))])
        for class_idx, class_name in enumerate(class_folders):
            class_dir = os.path.join(data_dir, class_name)
            class_image_files = []
        # Iterate over image files in the class folder
            for ext in ['*.jpg', '*.jpeg', '*.png', '*.bmp', '*.gif']:
                class_image_files.extend(tf.io.gfile.glob(os.path.join(class_dir, ext)))
         # Limit the number of samples for each class
            max_samples = min(len(class_image_files), 150)
            class_image_files = class_image_files[:max_samples]
          # Collect the image files and labels
            image_files.extend(class_image_files)
            labels.extend([class_idx] * len(class_image_files))
            class_names.append(class_name)

        print("Class Distribution:", {name: labels.count(i) for i, name in enumerate(class_names)})

    except Exception as e:
        print(f"Error collecting data: {e}")
        return [], [], []

    return image_files, labels, class_names
## Create dataset
def create_dataset(image_files, labels, preprocess_fn=None, is_training=True):# Function to create a TensorFlow dataset
    def process_path(file_path, label):# Function to process a single image
        try:
            img = tf.io.read_file(file_path)# Read the image file
            img = tf.image.decode_image(img, channels=3, expand_animations=False)# Decode the image
            img = tf.image.resize(img, [IMG_HEIGHT, IMG_WIDTH])# Resize the image

            if preprocess_fn is not None:
                img = preprocess_fn(img)
            else:
                img = tf.cast(img, tf.float32) / 255.0
            # Data Augmentation
            if is_training:
                img = tf.image.random_flip_left_right(img)
                img = tf.image.random_brightness(img, 0.2)
                img = tf.image.random_contrast(img, 0.8, 1.2)

        except Exception as e:
            print(f"Error processing image {file_path}: {e}")
            img = tf.zeros((IMG_HEIGHT, IMG_WIDTH, 3), dtype=tf.float32)

        return img, tf.one_hot(label, NUM_CLASSES)
     # Create the dataset
    dataset = tf.data.Dataset.from_tensor_slices((image_files, labels))
    if is_training:
        dataset = dataset.shuffle(buffer_size=len(image_files))
    dataset = dataset.map(process_path, num_parallel_calls=AUTO)
    dataset = dataset.batch(BATCH_SIZE)
    dataset = dataset.prefetch(AUTO)
    return dataset
## Create simple CNN
def create_simple_cnn():# Function to create a simple CNN
    model = models.Sequential([
        layers.Input(shape=(IMG_HEIGHT, IMG_WIDTH, 3)),# Input layer

        # First Convolutional Block
        layers.Conv2D(32, 3, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(32, 3, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(),
        layers.Dropout(0.25),

        # Second Convolutional Block
        layers.Conv2D(64, 3, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(64, 3, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(),
        layers.Dropout(0.25),

        # Third Convolutional Block
        layers.Conv2D(128, 3, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(128, 3, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(),
        layers.Dropout(0.25),

        # Dense Layers
        layers.Flatten(),
        layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.01)),# Regularization
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(NUM_CLASSES, activation='softmax')# Output layer
    ])
    return model
## Create VGG model
def create_vgg_model():# Function to create a VGG model
    base_model = VGG16(weights='imagenet', include_top=False, input_shape=(IMG_HEIGHT, IMG_WIDTH, 3))
    base_model.trainable = False

    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(NUM_CLASSES, activation='softmax')
    ])
    return model

def create_resnet_model():# Function to create a ResNet model
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(IMG_HEIGHT, IMG_WIDTH, 3))
    base_model.trainable = False

    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)),# Regularization
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.001)),# Regularization
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(NUM_CLASSES, activation='softmax')
    ])
    return model
## Train model
def train_model(model, train_dataset, test_dataset, train_labels, model_name):# Function to train a model
    class_weights = compute_class_weight(
        'balanced',
        classes=np.unique(train_labels),
        y=train_labels
    )# Compute class weights
    class_weight_dict = dict(enumerate(class_weights)) # Convert class weights to a dictionary

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
        loss=focal_loss(gamma=2., alpha=0.25),
        metrics=['accuracy',
                tf.keras.metrics.Precision(name='precision'),
                tf.keras.metrics.Recall(name='recall'),
                tf.keras.metrics.AUC(name='auc')]
    )# Compile the model

    callbacks = [
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=7,
            restore_best_weights=True
        ),
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=3,
            min_lr=1e-6
        )
    ]# Define callbacks

    try:
        history = model.fit(
            train_dataset,
            epochs=EPOCHS,
            validation_data=test_dataset,
            class_weight=class_weight_dict,
            callbacks=callbacks
        )# Train the model
        print(f"\nTraining completed for {model_name}")
        return history
    except Exception as e:
        print(f"Training error for {model_name}: {e}")
        return None
## Plot comparison
def plot_comparison(histories, model_names):# Function to plot model comparison
    for idx, (history, name) in enumerate(zip(histories, model_names)):
        # Plot Accuracy
        plt.figure(figsize=(8, 6))
        plt.plot(history.history['accuracy'], label='Training Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.title(f'{name} - Accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.show()

        # Plot Loss
        plt.figure(figsize=(8, 6))
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title(f'{name} - Loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()
        plt.show()

def plot_confusion_matrix(model, test_dataset, class_names):# Function to plot confusion matrix
    y_true = []
    y_pred = []

    for images, labels in test_dataset:
        y_true.extend(np.argmax(labels.numpy(), axis=1))
        y_pred.extend(np.argmax(model.predict(images), axis=1))

    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)# Plot confusion matrix
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

def main():
    data_dir = 'ECG_Image_data/train'

    # Collect balanced dataset
    image_files, labels, class_names = balanced_data_collection(data_dir)

    if not image_files:
        print("No images found. Check your data directory.")
        return

    # Split data
    train_files, test_files, train_labels, test_labels = train_test_split(
        image_files, labels,
        test_size=0.2,
        stratify=labels,
        random_state=42
    )

    # Create datasets for each model
    train_dataset_simple = create_dataset(train_files, train_labels, None, True)
    test_dataset_simple = create_dataset(test_files, test_labels, None, False)

    train_dataset_vgg = create_dataset(train_files, train_labels, vgg_preprocess, True)
    test_dataset_vgg = create_dataset(test_files, test_labels, vgg_preprocess, False)

    train_dataset_resnet = create_dataset(train_files, train_labels, resnet_preprocess, True)
    test_dataset_resnet = create_dataset(test_files, test_labels, resnet_preprocess, False)

    # Train Simple CNN
    print("\nTraining Simple CNN...")
    simple_cnn = create_simple_cnn()
    simple_history = train_model(simple_cnn, train_dataset_simple, test_dataset_simple, train_labels, "Simple CNN")

    # Train VGG16 Model
    print("\nTraining VGG16...")
    vgg_model = create_vgg_model()
    vgg_history = train_model(vgg_model, train_dataset_vgg, test_dataset_vgg, train_labels, "VGG16")

    # Train ResNet50 Model
    print("\nTraining ResNet50...")
    resnet_model = create_resnet_model()
    resnet_history = train_model(resnet_model, train_dataset_resnet, test_dataset_resnet, train_labels, "ResNet50")

    # Plot comparisons
    plot_comparison([simple_history, vgg_history, resnet_history], ["Simple CNN", "VGG16", "ResNet50"])

    # Plot confusion matrices
    plot_confusion_matrix(simple_cnn, test_dataset_simple, class_names)
    plot_confusion_matrix(vgg_model, test_dataset_vgg, class_names)
    plot_confusion_matrix(resnet_model, test_dataset_resnet, class_names)

if __name__ == "__main__":
    main()
```
